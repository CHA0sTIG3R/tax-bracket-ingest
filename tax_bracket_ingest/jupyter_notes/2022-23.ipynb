```python
def process_tax_brackets(df, year, single_filer_col):
    # Remove header/footer rows and set correct headers
    df = df.copy()
    df.columns = df.iloc[0]
    df = df[1:]
    df = df[:-1]
    
    # Extract range starts for each status
    s_range_start = df[single_filer_col].str.split(' ', expand=True)[0].str.strip()
    mfj_range_start = df['For Married Individuals Filing Joint Returns'].str.split(' ', expand=True)[0].str.strip()
    hoh_range_start = df['For Heads of Households'].str.split(' ', expand=True)[0].str.strip()
    
    status_rates = [
        "Married Filing Jointly (Rates/Brackets)",
        "Married Filing Separately (Rates/Brackets)",
        "Single Filer (Rates/Brackets)",
        "Head of Household (Rates/Brackets)"
    ]
    greater_than = '>'
    rates = df[df.columns[0]] * 100
    rates = rates.astype(str) + '%'
    
    rows = []
    for status, range_start in zip(status_rates, [mfj_range_start, s_range_start, s_range_start, hoh_range_start]):
        sub_row = rates.to_frame().copy()
        sub_row.insert(0, 'Year', year)
        sub_row.insert(2, 'For Income >', greater_than)
        sub_row.insert(3, 'Range Start', range_start)
        sub_row.rename(columns={sub_row.columns[1]: status}, inplace=True)
        rows.append(sub_row.reset_index(drop=True))
    
    merged_df = pd.concat(rows, axis=1)
    # Remove duplicate 'Year' columns
    cols = merged_df.columns.tolist()
    seen = {}
    new_cols = []
    for col in cols:
        if col == 'Year':
            count = seen.get(col, 0)
            if count == 0:
                new_cols.append(True)
            else:
                new_cols.append(False)
            seen[col] = count + 1
        else:
            new_cols.append(True)
    merged_df = merged_df.loc[:, new_cols]
    return merged_df
```

```python
prcessed_22 = process_tax_brackets(df_22, 2022, 'For Unmarried Individuals')
prcessed_22
```

```python
processed_23 = process_tax_brackets(df_23, 2023, 'For Single\xa0Filers')
processed_23
```

```python
combined = pd.concat([prcessed_22, processed_23], ignore_index=True)
combined
```

```python
combined.sort_values(
    by=['Year', 'Married Filing Jointly (Rates/Brackets)'],
    ascending=[False, True],
    inplace=True
)
combined
```

```python
combined.to_csv("../data/normalized/2022-23-tax-brackets.csv", index=False)
print("Tax brackets processed and saved to ../data/normalized/2022-23-tax-brackets.csv")
```
