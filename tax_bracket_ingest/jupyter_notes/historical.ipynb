```python
from datetime import datetime
from dotenv import load_dotenv
import os
import pandas as pd
import numpy as np
```

```python
load_dotenv()

data_path = os.getenv("DATA_PATH")

historical_data = pd.read_csv(data_path +'/raw/Historical Income Tax Rates and Brackets, 1862-2021.csv')
historical_data
```

```python
# remove null rows
historical_data = historical_data.dropna(how='all')
historical_data
```

```python
# Rename the unnamed columns
greater_than = 'For Income >'
range_start = 'Range Start'
# Automatically rename all 'Unnamed: x' columns to the appropriate label in a DRY way
rename_map = {}
for i, col in enumerate(historical_data.columns):
    if 'Unnamed' in col:
        # Odd-numbered unnamed columns are 'For Income >', even-numbered are 'Range Start'
        rename_map[col] = greater_than if i % 3 != 0 else range_start

historical_data = historical_data.rename(columns=rename_map)

historical_data
```

```python
historical_data.dtypes
```

```python
historical_data['Year'] = historical_data['Year'].replace('1940(A)', '1940')
historical_data['Year'] = historical_data['Year'].astype(int)
historical_data.dtypes
```

```python
df_22_23 = pd.read_csv(data_path+'/normalized/2022-23-tax-brackets.csv')
df_22_23
```

```python
df_22_23.columns = df_22_23.columns.str.replace(r'\.\d+$', '', regex=True)
df_22_23
```

```python
df_22_23.dtypes
```

```python
notes = historical_data['Notes:']
notes
```

```python
historical_data = historical_data.drop(columns=['Notes:'])
historical_data
```

```python
updated_data = pd.concat([historical_data, df_22_23], ignore_index=True)
updated_data
```

```python
# Remove all "For Income >" columns
updated_data = updated_data.drop('For Income >', axis=1)
updated_data
```

```python
new_names = ['MFJ Range Start', 'MFS Range Start', 'S Range Start', 'HOH Range Start']

for i in range(1, len(new_names)+1):
    updated_data.columns.values[i*2] = new_names[i-1]
    
updated_data
```

```python
updated_data.sort_values(
    by=['Year', 'Married Filing Jointly (Rates/Brackets)'],
    ascending=[False, True],
    inplace=True
)
updated_data
```

```python
final_df = updated_data.copy()

group_col = "Year"

range_start_cols = [col for col in final_df.columns if "Range Start" in col]

for col in range_start_cols:
    
    numeric_start = (
        final_df[col]
        .astype(str)
        .str.replace(r'[^\d.]', '', regex=True)
        .replace(r'^\s*$', np.nan, regex=True)
        .astype(float)
    )
    
    numeric_end = (
        numeric_start
        .groupby(final_df[group_col], sort=False)
        .apply(lambda grp: grp.shift(-1) - 1)
        .reset_index(level=0, drop=True)
    )
    
    formatted_end = numeric_end.apply(
        lambda x: f"${int(x):,}" if pd.notnull(x) else np.nan
    )

    ins_idx = final_df.columns.get_loc(col) + 1
    final_df.insert(ins_idx, col.replace("Start", "End"), formatted_end)

final_df
```

```python
final_df.to_csv(data_path +'/normalized/normalized-historical-tax-brackets.csv', index=False)
print(f"Combined tax brackets data saved to '../data/normalized/combined-tax-brackets.csv' at {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
```
