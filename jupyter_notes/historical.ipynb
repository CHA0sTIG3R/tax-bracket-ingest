```python
import sys
import os
from datetime import datetime
import pandas as pd

# Add the parent directory to sys.path so local modules can be imported
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
```

```python
historical_data = pd.read_csv('../data/raw/Historical Income Tax Rates and Brackets, 1862-2021.csv')
historical_data
```

```python
# remove null rows
historical_data = historical_data.dropna(how='all')
historical_data
```

```python
# Rename the unnamed columns
greater_than = 'For Income >'
range_start = 'Range Start'
# Automatically rename all 'Unnamed: x' columns to the appropriate label in a DRY way
rename_map = {}
for i, col in enumerate(historical_data.columns):
    if 'Unnamed' in col:
        # Odd-numbered unnamed columns are 'For Income >', even-numbered are 'Range Start'
        rename_map[col] = greater_than if i % 3 != 0 else range_start

historical_data = historical_data.rename(columns=rename_map)

historical_data
```

```python
historical_data.dtypes
```

```python
historical_data['Year'] = historical_data['Year'].replace('1940(A)', '1940')
historical_data['Year'] = historical_data['Year'].astype(int)
historical_data.dtypes
```

```python
df_22_23 = pd.read_csv('../data/normalized/2022-23-tax-brackets.csv')
df_22_23
```

```python
df_22_23.columns = df_22_23.columns.str.replace(r'\.\d+$', '', regex=True)
df_22_23
```

```python
df_22_23.dtypes
```

```python
notes = historical_data['Notes:']
notes
```

```python
historical_data = historical_data.drop(columns=['Notes:'])
historical_data
```

```python
updated_data = pd.concat([historical_data, df_22_23], ignore_index=True)
updated_data
```

```python
updated_data.sort_values(
    by=['Year', 'Married Filing Jointly (Rates/Brackets)'],
    ascending=[False, True],
    inplace=True
)
updated_data
```

```python
updated_data.to_csv('../data/normalized/normalized-historical-tax-brackets.csv', index=False)
print(f"Combined tax brackets data saved to '../data/normalized/combined-tax-brackets.csv' at {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
```
